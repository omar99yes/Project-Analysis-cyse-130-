Task 2: Automating a Security Check (10 Points)

Write a Python function to check if a list of servers is online. The function should:

    Accept a list of IP addresses as input.
    Ping each server using the os.system command.
    Print "Online" or "Offline" for each IP address.
    
Start with this template:

```python
import os

def check_servers(ip_list):
    for ip in ip_list:
        response = os.system(f"ping -c 1 {ip}")
        if response == 0:
            print(f"{ip} is Online")
        else:
            print(f"{ip} is Offline")

# Example usage
servers = ["192.168.1.1", "8.8.8.8", "10.0.0.5"]
check_servers(servers)
```

Task 3: Log File Analysis (10 Points)

Write a Python script to count the number of times the word "ERROR" appears in a given log file.

Instructions:

    Use the open() function to read the file.
    Loop through each line of the file and check if "ERROR" is in the line.
    Maintain a counter to track the number of occurrences.
    Print the total count at the end.
    
Start with this template:

```python
def count_errors(log_file):
    error_count = 0

    # Open the file in read mode
    with open(log_file, "r") as file:
        for line in file:
            if "ERROR" in line:
                error_count += 1

    print(f"Total number of errors: {error_count}")

# Example usage
log_file_path = "system_logs.txt"  # Replace with your log file name
count_errors(log_file_path)
```
